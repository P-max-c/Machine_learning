```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv(r"C:\Users\hegde\Downloads\Ice_cream selling data.csv")

# Create scatter plot using the correct column name
plt.scatter(dataset["Temperature (°C)"], dataset["Ice Cream Sales (units)"])
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.title("Ice Cream Sales vs Temperature")
plt.show()
```

```python
x=dataset[["Temperature (°C)"]]
y=dataset["Ice Cream Sales (units)"]
```

```python
# Import necessary libraries
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Create polynomial features
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and fit the model
lr = LinearRegression()  # Need parentheses to create an instance
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
pd=lr.predict(x)
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.plot(dataset["Temperature (°C)"],pd)
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```
