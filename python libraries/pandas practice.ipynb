```python
import pandas as p
# series using pandas
d=[2,2.1,"a"]
s=p.Series(d, index=["a","b","c"])
s.loc["c"]=20000
print(s[s<200])
```

```python
c={"d 1":1750,"d 2":2100,"d 3":1700}
s=p.Series(c)
s.loc["d 3"]+=500
print(s[s<2000])
```
